```python
import pandas
```

```python
lego = pandas.read_csv('../data/lego_sets.csv')
theme = pandas.read_csv('../data/themes.csv')
sets = pandas.read_csv('../data/sets.csv')
parent_theme = pandas.read_csv('../data/parent_themes.csv')
```

```python
parent_theme.head()
```

```python
lego.head()
```

```python
merges = pandas.merge(lego, parent_theme, left_on="parent_theme", right_on="name")
```

```python
merges.drop(columns="name_y", inplace=True)
```

```python
merges.rename(columns={"name_x": "name"}, inplace=True)
```

```python
licensed = merges[merges["is_licensed"]]
licensed = licensed.dropna(subset=["set_num"])

star_wars = licensed[licensed["parent_theme"] == "Star Wars"]

the_force = int(star_wars.shape[0] / licensed.shape[0] * 100)
```

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1
# licensed_sorted.head()

summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()

summed_df.head()
```
