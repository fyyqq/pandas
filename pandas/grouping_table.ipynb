```python
business = pandas.DataFrame({
    'companies' : ['Petronas', 'BHP', 'Maxis', 'MCD', 'Tesla', 'MCD', 'Repace', 'Repace'],
    'employees': ['Sarah', 'Fyqq', 'Alisha', 'Azim', 'Dasha', 'Intan', 'Fizi', 'Elon'],
    'ages': [22, 55, 25, 44, 32, 65, 19, 37]
})
```

```python
business
```

```python
business.groupby('ages').sum()
```

```python
business.groupby('companies').count()
```

```python
business.groupby('ages').min()
```

```python
business.groupby('ages').ages.min()
```

```python
business.groupby('companies').describe().transpose()
```

```python
business.groupby('ages').describe().transpose()
```

```python
data1 = pandas.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3', 'k4'],
    'one': [1, 2, 3, 4, 5],
    'two': [6, 7, 8, 9, 10]
})
```

```python
data2 = pandas.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3', 'k4'],
    'three': [1, 2, 3, 4, 5],
    'four': [6, 7, 8, 9, 10]
})
```

```python
data1
```

```python
data2
```

```python
pandas.merge(data1, data2, how='inner', on='key')
```

```python
x =  pandas.DataFrame({
    'key1' : ['K0', 'K0', 'K1', 'K2'],
    'key2' : ['K0', 'K1', 'K0', 'K1'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3'],
})

y =  pandas.DataFrame({
    'key1' : ['K0', 'K0', 'K1', 'K2'],
    'key2' : ['K0', 'K1', 'K0', 'K1'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3'],
})
```

```python
x
```

```python
y
```

```python
pandas.merge(x, y, how='left', on=['key1', 'key2'])
```

```python
pandas.merge(x, y, how='right', on=['key1', 'key2'])
```

```python
dad = pandas.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

mom = pandas.DataFrame({
    'C': ['C0', 'C2', 'C3'],
    'D': ['D0', 'D2', 'D3']
}, index=['K0', 'K2', 'K3'])
```

```python
dad
```

```python
mom
```

```python
# JOIN

dad.join(mom)
```

```python
mom.join(dad)
```

```python
dad.join(mom, how='inner')
```
