```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame contoh
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 30, 20, 25]}

df = pd.DataFrame(data)

# Menghasilkan bar chart
df.plot(kind='bar', x='Category', y='Values', legend=False)
plt.title('Bar Chart')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame contoh
data = {'X': [1, 2, 3, 4],
        'Y': [10, 20, 25, 30]}

df = pd.DataFrame(data)

# Menghasilkan line chart
df.plot(kind='line', x='X', y='Y', marker='o')
plt.title('Line Chart')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame contoh
data = {'X': [1, 2, 3, 4],
        'Y': [10, 20, 25, 30]}

df = pd.DataFrame(data)

# Menghasilkan scatter plot
df.plot(kind='scatter', x='X', y='Y', color='red')
plt.title('Scatter Plot')
plt.show()
```
